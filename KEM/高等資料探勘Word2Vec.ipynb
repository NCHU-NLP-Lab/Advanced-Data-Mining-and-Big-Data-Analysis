#### Similar Function

```python
from gensim import models

model = models.Word2Vec.load('Your model here')

def sim(input):
    """
    Return 10 most related keywords from word2vec result.
    Parameters
    ----------
    input : string
        only keywords on word2vec dict will return result, otherwise return empty list
    Returns
    -------
    value : list of dict, result keywords as keys, and its frequency will be value.
    """
    res = []
    if len(input) == 0:
        print('none')
    else:
        try:
            
            for w2v in model.wv.most_similar(input, topn = 10):
                print(w2v)
                res.append(w2v)

            print(res)
            
        except Exception as e:
            print('not found keyword')
            
    return res
```
